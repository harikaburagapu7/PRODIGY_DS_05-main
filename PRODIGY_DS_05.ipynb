```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('US_Accidents_March23.csv')

# Aggregate the number of accidents by state
accidents_by_state = data['State'].value_counts()

# Create a bar graph
plt.figure(figsize=(12, 6))
accidents_by_state.plot(kind='bar', color='purple')
plt.title('Number of Accidents per State')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('US_Accidents_March23.csv')

# Check if 'Severity' column exists
if 'Severity' in data.columns:
    # Create a histogram for the 'Severity' column
    plt.figure(figsize=(10, 6))
    plt.hist(data['Severity'], bins=range(1, 7), edgecolor='black', color='skyblue', rwidth=0.8)
    plt.title('Distribution of Accident Severity')
    plt.xlabel('Severity')
    plt.ylabel('Frequency')
    plt.xticks(range(1, 6))  # Assuming severity ranges from 1 to 5
    plt.show()
else:
    print("The column 'Severity' does not exist in the dataset.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style="whitegrid")

# Plot 1: Histogram of Accident Severity
plt.figure(figsize=(10, 6))
sns.histplot(data['Severity'], bins=10, kde=False, color='blue')
plt.title('Histogram of Accident Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 2: Bar Chart of Accidents by State
plt.figure(figsize=(12, 8))
state_counts = data['State'].value_counts()
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.title('Number of Accidents by State')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot 3: Scatter Plot of Accidents by Latitude and Longitude
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Start_Lng'], y=data['Start_Lat'], alpha=0.1)
plt.title('Scatter Plot of Accidents by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Plot 4: Line Chart of Accidents over Time
data['Start_Time'] = pd.to_datetime(data['Start_Time'])
data.set_index('Start_Time', inplace=True)
accidents_over_time = data.resample('M').size()
plt.figure(figsize=(12, 6))
accidents_over_time.plot()
plt.title('Accidents over Time')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Plot 5: Bar Chart of Accidents by Weather Condition
plt.figure(figsize=(12, 8))
weather_counts = data['Weather_Condition'].value_counts().nlargest(10)
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette='coolwarm')
plt.title('Top 10 Weather Conditions for Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python

# Plot 6: Heatmap of Accidents by Hour and Day of the Week
data['Hour'] = data.index.hour
data['Day_of_Week'] = data.index.dayofweek
hour_day_pivot = data.pivot_table(index='Day_of_Week', columns='Hour', values='ID', aggfunc='count')
plt.figure(figsize=(14, 8))
sns.heatmap(hour_day_pivot, cmap='YlGnBu')
plt.title('Heatmap of Accidents by Hour and Day of the Week')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
# Plot 7: Box Plot of Accident Severity by Weather Condition
plt.figure(figsize=(14, 8))
top_weather_conditions = data['Weather_Condition'].value_counts().nlargest(10).index
sns.boxplot(x='Weather_Condition', y='Severity', data=data[data['Weather_Condition'].isin(top_weather_conditions)])
plt.title('Box Plot of Accident Severity by Top Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 8: Pie Chart of Accidents by Road Type
plt.figure(figsize=(8, 8))
road_type_counts = data['Road_Type'].value_counts()
plt.pie(road_type_counts, labels=road_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart of Accidents by Road Type')
plt.show()
```

```python
# Plot 9: Histogram of Accidents by Temperature
plt.figure(figsize=(10, 6))
sns.histplot(data['Temperature(F)'], bins=30, kde=True, color='red')
plt.title('Histogram of Accidents by Temperature')
plt.xlabel('Temperature (F)')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 10: Bar Chart of Accidents by City
plt.figure(figsize=(14, 8))
city_counts = data['City'].value_counts().nlargest(10)
sns.barplot(x=city_counts.index, y=city_counts.values, palette='Set2')
plt.title('Top 10 Cities with Most Accidents')
plt.xlabel('City')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```
